```python
import tensorflow as tf
import pandas as pd
import numpy as np
import os
```

```python
df=pd.read_csv(os.path.join('datasets','spam.csv'), encoding='ISO-8859-1')
```

```python
df.head()
```

```python
df=df[df.columns[:2]]
```

```python
df.head()
```

```python
df.shape
```

```python
texts=df['v2'].values
```

```python
texts
```

```python
from tensorflow.keras.layers import TextVectorization
```

```python
MAX_FEATURES=200000
```

```python
vectorizer=TextVectorization(max_tokens=MAX_FEATURES,output_sequence_length=1800,output_mode='int')
```

```python
vectorizer.adapt(texts)
```

```python
len(vectorizer.get_vocabulary())
```

```python
vectorized_text=vectorizer(texts)
```

```python
vectorized_text
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer=Tokenizer(num_words=200000)
tokenizer.fit_on_texts(texts)
```

```python
sequences=tokenizer.texts_to_sequences(texts)
```

```python
sequences
```

```python
sequences=tf.ragged.constant(sequences)
```

```python
sequences
```

```python
y=df['v1'].map({'ham':0,'spam':1})
```

```python
y=y.values
```

```python
# y=np.expand_dims(y,0)
```

```python
y
```

```python
#MCSHBAP
dataset=tf.data.Dataset.from_tensor_slices((vectorized_text,y))
dataset=dataset.cache()
dataset=dataset.shuffle(100)
dataset=dataset.batch(16)
dataset=dataset.prefetch(8)
```

```python
len(dataset)
```

```python
train=dataset.take(int(len(dataset)*.7))
val=dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test=dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Bidirectional,LSTM,Embedding
```

```python
model=Sequential()
model.add(Embedding(MAX_FEATURES+1,32))
model.add(Bidirectional(LSTM(32,activation='tanh')))
model.add(Dense(128,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
history=model.fit(train,epochs=8,validation_data=val)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
pd.DataFrame(history.history).plot()
plt.show()
```

```python
from tensorflow.keras.metrics import Precision,Recall,BinaryAccuracy
```

```python
pre=Precision()
rec=Recall()
ba=BinaryAccuracy()
```

```python
x_1,y_1=test.as_numpy_iterator().next()
```

```python
y_1.flatten()
```

```python
model=tf.keras.models.load_model('saved_models/spamD.h5')
```

```python
for batch in test.as_numpy_iterator():
    X_true,y_true=batch
    yhat=model.predict(X_true)
    yhat=(yhat>=0.5).astype(int)
    yhat=yhat.flatten()
    y_true=y_true.flatten()
    
    # print(accuracy_score(y_true,yhat))
    
    pre.update_state(yhat,y_true)
    rec.update_state(yhat,y_true)
    ba.update_state(yhat,y_true)
```

```python
print(f'Precision {pre.result().numpy()}, Recall {rec.result().numpy()}, Accuracy {ba.result().numpy()}')
```

### Different Method

```python
dataset2=tf.data.Dataset.from_tensor_slices((sequences,y))
dataset2=dataset2.cache()
dataset2=dataset2.shuffle(100)
dataset2=dataset2.batch(16)
dataset2=dataset2.prefetch(8)
```

```python
len(dataset2)
```

```python
train2=dataset2.take(int(len(dataset2)*.7))
val2=dataset2.skip(int(len(dataset2)*.7)).take(int(len(dataset2)*.2))
test2=dataset2.skip(int(len(dataset2)*.9)).take(int(len(dataset2)*.1))
```

```python
model2=Sequential()
model2.add(Embedding(MAX_FEATURES+1,32))
model2.add(Bidirectional(LSTM(32,activation='tanh')))
model2.add(Dense(128,activation='relu'))
model2.add(Dense(256,activation='relu'))
model2.add(Dense(128,activation='relu'))
model2.add(Dense(1,activation='sigmoid'))
```

```python
model2.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model2.summary()
```

```python
history2=model2.fit(train2,epochs=8,validation_data=val2)
```

```python
plt.figure(figsize=(8,5))
pd.DataFrame(history2.history).plot()
plt.show()
```

```python
for batch2 in iter(test2):
    X_true,y_true=batch2
    yhat=model2.predict(X_true)
    yhat=(yhat>=0.5).astype(int)
    yhat=yhat.flatten()
    # print(y_true)
    y_true=y_true.numpy()
    y_true=y_true.flatten()
    
    # print(accuracy_score(y_true,yhat))
    
    pre.update_state(yhat,y_true)
    rec.update_state(yhat,y_true)
    ba.update_state(yhat,y_true)
```

```python
print(f'Precision {pre.result().numpy()}, Recall {rec.result().numpy()}, Accuracy {ba.result().numpy()}')
```

#### conclusion: Tokenizer and TextVectorization gives the same accuracy but Tokenizer helps in training the network faster
